# Quick EDA: Credit Card Fraud Data

```
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
df = pd.read_csv('data/creditcard.csv')
print(df.shape)
df.head()
```

```
print(Counter(df['Class']))
df['Amount'].describe() if 'Amount' in df.columns else None
```

```
# Class balance bar
counts = df['Class'].value_counts().sort_index()
counts.plot(kind='bar')
plt.title('Class Balance (0=non-fraud, 1=fraud)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```
